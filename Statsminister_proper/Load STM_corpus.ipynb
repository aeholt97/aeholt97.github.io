### Below are some further manipulation of the data, coding it for party and prime-minister

```python
number_of_failures = 0
Y = []
M = []
D = []
for i, s in enumerate(data['date']):
    if s != "Nan":
        d,m,y = s.split('.')
        Y.append(y)
        M.append(m)
        D.append(d)
    else:
        Y.append('NaN')
        M.append('NaN')
        D.append('NaN')
        number_of_failures +=1
print('Done!\n',number_of_failures,'links failed')
```

```python
for i,y in enumerate(Y):
    if y != "NaN":
        if int(y) < 30:
            Y[i] = "20"+y
        else:
            Y[i] = "19"+y
    else:
        y=y
```

```python
data['DD'] = D
data['MM'] = M
data['YYYY'] = Y

data
```

```python
def get_party(y,m,d):
    if str(m)[0] == "0":
        m = str(m).replace("0","")
    if str(d)[0] == "0":
        d = str(d).replace("0","")
    y = int(y)
    m = int(m)
    d = int(d)
    if y <= 2001:
        party = "S"
        pm = "Nyrup"
        if y == 2001 and m > 11:
            party = "V"
            pm = "Fogh"
    elif y <= 2009:
        party = "V"
        pm = "Fogh"
        if y == 2009 and m > 3:
            party = "V"
            pm = "Løkke 1"
    elif y <= 2011:
        party = "V"
        pm = "Løkke 1"
        if y == 2011 and m > 9:
            party = "S"
            pm = "Thorning"
    elif y <= 2015:
        party = "S"
        pm = "Thorning"
        if y == 2015 and m > 6:
            party = "V"
            pm = "Løkke 2"
    elif y <= 2019:
        party = "V"
        pm = "Løkke 2"
        if y == 2019 and m > 6:
            party = "S"
            pm = "Frederiksen"
    elif y > 2019:
        party = "S"
        pm = "Frederiksen"
    else:
        party = "X"
        pm = "X"
    return party, pm
```

```python
party_list = []
pm_list = []
for i in range(0,len(data['date']),1):
    if data['date'][i] != "Nan":
        party_list.append(get_party(data['YYYY'][i],data['MM'][i],data['DD'][i])[0])
        pm_list.append(get_party(data['YYYY'][i],data['MM'][i],data['DD'][i])[1])
    else:
        party_list.append('NaN')
        pm_list.append('NaN')
```

```python
pm_list
```

```python
data['pm'] = pm_list
data['party'] = party_list
data
```
