### The function url_obtain, defined below, takes two arguments; HTML-class and type of text
#### p_class refers to the HTML-class of the /div (eg. 'component richtext', 'article-top')
#### p_type refers to the type of text-tag (eg. 'p', 'h1')

```python
# First we import modules for HTML manipulation
import requests
from bs4 import BeautifulSoup
```

```python
def url_obtain(url, p_class, p_type):
    try:
        '''Returns text data from website'''
        page = requests.get(url).text
        soup = BeautifulSoup(page, "lxml")
        text = [p.text for p in soup.find(class_=p_class).find_all(p_type)]
    #    print("found: " + p_type + "   in class: " + p_class + "   for link ending in: " + str(url[-10:]))
        return text
    except:
        return 'NaN'
```

### Below is the main scraping loop. It is a bit complicated and takes a long time to run
#### Make sure to not delete the sleep() function as you will otherwise be charged with carrying out a DDOS attack against the Prime Ministry

```python
# First we define the variable urls as the list of urls in the dataframe
urls = url_corpus['link']
from os import system

from time import sleep
from random import randint

# We are interested in three different things from each speech:
titles = []
transcripts = []
dates = []

# This loop uses the function url_obtain to get dates, text and title
# from urls
for i, url in enumerate(urls):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    titles.append(url_obtain(url, 'article-top', 'h1'))
    transcripts.append(url_obtain(url, 'component richtext', 'p'))
    sleep(randint(1,4))
    print("Finished link " + str(i+1) + " of " + str(len(urls)))
    try:
        dates.append(soup.time.text)
    except:
        dates.append('Nan')
        
# after running the loop we combine all the data in a DataFrame
import pandas as pd
data = pd.DataFrame({'date': dates,
                     'title': titles,
                     'speech': transcripts })
data
```

### The speeches are converted to single strings

```python
# Talerne har meget forskellige opdelinger af strings, derfor samler vi hver tale i en enkelt string

taler = data['speech']
for i, x in enumerate(taler):
    str_lists = x
    s=" "
    taler[i] = s.join(str_lists)
print(taler[1][0:100])
```

### Then we want to clean the speeches to make text-processing easier and add them to the DataFrame

```python
# A list is created to hold the speeches when they have been cleaned
cleaned_speeches = []

for string in taler:
    new_string = ""
    for letter in string:
        if letter not in ".”,:;-=–?+-#*/%!¤()=[]{}1234567890":
            new_string += str(letter.lower())
    cleaned_speeches.append(new_string)

data['speech_clean'] = cleaned_speeches
```

```python
# herunder anvendes funktionen fjern_stop_ord og talerne laves hver især om til en enkelt string igen.
stopped = []
for i, x in enumerate(corpus['speech_clean']):
    stopped.append(fjern_stop_ord(corpus['speech_clean'][i]))
```

```python
# We are going to create a document-term matrix using CountVectorizer, and exclude common English stop words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
data_cv = cv.fit_transform(corpus.speech_clean)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = corpus.index
data_dtm
```

```python
data = data_dtm

data = data.transpose()
data.head()

# Find the top 40 words said by each speech
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(20)
    top_dict[c]= list(zip(top.index, top.values))

# Ovenover kan du vælge hvilken top # der skal gemmes i top_dict (std = 40)
# I det nedenstående kan du ændre rangen for at se ord rangeret efter
# antallet af steder det forekommer i teksten.
pd_top = pd.DataFrame(top_dict)
pd_top
```
