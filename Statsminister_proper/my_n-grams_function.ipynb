```python
def create_ngrams_from_string(string,n):
    ngrams = []
    string = string.split(' ')
    for i in range(0,len(string),1):
        ngrams.append(string[i:i+n])
    for j, n  in enumerate(ngrams):
        s = ""
        for i in range(0,len(n), 1):
            s+=str(n[i])+" "
        ngrams[j] = s[:-1]
    return ngrams

def create_ngrams_from_list(list,n):
    ngrams = []
    for i in range(0,len(list),1):
        ngrams.append(list[i:i+n])
    for j, n  in enumerate(ngrams):
        s = ""
        for i in range(0,len(n), 1):
            s+=str(n[i])+" "
        ngrams[j] = s[:-1]
    return ngrams
```

```python
unigrams = []
for s in data['speech_clean']:
    unigrams += create_ngrams_from_string(s,1)
bigrams = []
for s in data['speech_clean']:
    bigrams += create_ngrams_from_string(s,2)
trigrams = []
for s in data['speech_clean']:
    trigrams += create_ngrams_from_string(s,3)


#unigrams = create_ngrams_from_string(data['speech_clean'][0],1)
#bigrams = create_ngrams_from_string(data['speech_clean'][0],2)
#trigrams = create_ngrams_from_string(data['speech_clean'][0],3)


uni_dic = {}
for u in unigrams:
    if u in uni_dic:
        uni_dic[u]+=1
    else:
        uni_dic[u]=1
uni_dic = dict(sorted(uni_dic.items(), key=lambda item: item[1], reverse = True))

bi_dic = {}
for b in bigrams:
    if b in bi_dic:
        bi_dic[b]+=1
    else:
        bi_dic[b]=1
bi_dic = dict(sorted(bi_dic.items(), key=lambda item: item[1], reverse = True))

tri_dic = {}
for t in trigrams:
    if t in tri_dic:
        tri_dic[t]+=1
    else:
        tri_dic[t]=1
tri_dic = dict(sorted(tri_dic.items(), key=lambda item: item[1], reverse = True))
```

```python
print(len(uni_dic))
print(len(bi_dic))
print(len(tri_dic))
```

```python
tri_clean = {}
for i, t in enumerate(tri_dic):
    if tri_dic[t] != 1:
        tri_clean[t] = tri_dic[t]

bi_clean = {}
for i, t in enumerate(bi_dic):
    if bi_dic[t] != 1:
        bi_clean[t] = bi_dic[t]

uni_clean = {}
for i, t in enumerate(uni_dic):
    if uni_dic[t] != 1:
        uni_clean[t] = uni_dic[t]
```
